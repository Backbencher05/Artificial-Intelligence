```python
# Line Plot
```

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5]
```

```python
sales = [10,15,20,18,25]
```

```python
plt.plot(days, sales)
```

```python
plt.show()
```

```python
plt.title("Sales Trend")
```

```python
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```

```python
plt.plot(days, sales)
```

```python
plt.plot(days, sales)

plt.title("Sales Trend")

plt.xlabel("Day")

plt.ylabel("Sales")

plt.show()
```

```python
# Bar Chart: Used for categories.
```

```python
products = ["A", "B", "C"]

sales = [100, 150, 90]
```

```python
plt.bar(products, sales)
plt.show()
```
