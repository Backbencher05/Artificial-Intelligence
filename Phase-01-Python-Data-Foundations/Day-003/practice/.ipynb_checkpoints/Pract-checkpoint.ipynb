```python
# Part 5: Matrix Multiplication
```

```python
# Input Matrix
x = [
    [1,2],
    [3,4]
]
```

```python
# Weight Matrix
w = [
    [5,6],
    [7,8]
]
```

```python
# operation : x@w
# or
np.matmul(x,w)
```

```python
What Happened?

Row from first matrix

×

Column from second matrix

↓

Dot Product

This is the key.

Matrix multiplication is basically:

Many Dot Products

performed simultaneously.
```

```python
1*5+2*7 , 1*6+2*8
3*5+4*7 , 3*6+4*8 

5+14, 6+16
15+28, 18+32

# o/p
[[19, 22],
 [43, 50]]
```

```python
X = np.array([
    [1,2],
    [3,4]
])

W = np.array([
    [5,6],
    [7,8]
])

print(X @ W)
```

```python
print(X.shape)
print(W.shape)
```
