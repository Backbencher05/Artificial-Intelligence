```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Name": ["Aditya", "Rahul", "Priya", "Priya"],
    "Age": [28,30,25,25],
    "Salary": [100000, np.nan, 90000, 90000],
    "Department": ["IT", "HR", "IT", "IT"]
})
```

```python
print("Original Data")
print(df)
```

```python
print("Missing Values")
print(df.isnull().sum())
```

```python
print("Duplicates")
print(df.duplicated())
```

```python
print("After Removing Duplicates")
print(df.drop_duplicates())
```

```python
print("Fill Missing Salary with Mean")
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
```
